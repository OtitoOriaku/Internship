```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Send a GET request to the website
url = "https://www.dineout.co.in/delhi-restaurants"
response = requests.get(url)


response
```

```python
# Create a BeautifulSoup object with the response content
soup = BeautifulSoup(response.content, "html.parser")

soup
```

```python
# Find all the restaurant details
restaurants = soup.find_all("div", class_="restnt-info")

restaurants
```

```python
# Extract the details for each restaurant
data = []
for restaurant in restaurants:
    name = restaurant.find("a", class_="restnt-name").text.strip()
    cuisine = restaurant.find("span", class_="double-line-ellipsis").text.strip()
    location = restaurant.find("span", class_="double-line-ellipsis grey-text").text.strip()
    ratings = restaurant.find("span", class_="rating-val").text.strip()
    image_url = restaurant.find("img")["data-original"]
    data.append([name, cuisine, location, ratings, image_url])
```

```python
# Create a DataFrame with the extracted data
df = pd.DataFrame(data, columns=["Restaurant Name", "Cuisine", "Location", "Ratings", "Image URL"])

# Display the DataFrame
print(df)
```
